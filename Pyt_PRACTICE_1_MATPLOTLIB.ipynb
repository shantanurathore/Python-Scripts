```python
%matplotlib inline
from pylab import *

import matplotlib.pyplot as plt
import numpy as np

x=np.linspace(0,25,10) #the 10 here is probably log to the base 10
y=x**2

figure()
plot(x,y,'r') #r stands for color red g is green b is blue
xlabel('x')
ylabel('y')
title('square function')
show()

```

```python
#subplots -- what are subplots? I think they are layout managers in matplotlib
subplot(1,2,1)
plot(x, y, 'r--') # r-- means red line of type ----
subplot(1,2,2)
plot(y, x, 'g*-'); # g*- means green line with * and - 
```

```python
#defining axes - location of the axes for a plot
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height - values are between 0 and 1
#0.1 is basically start of the plot
axes.plot(x,y,'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('TITLE')
```

```python
#example of adding multiple axes to the plot - two graphs in a single space
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])#left, bottom, width, height - values are between 0 and 1
#0.1 is basically start of the plot
axes2 = fig.add_axes([0.15,0.39,0.4,0.4])

axes1.plot(x,y,'r')
axes2.plot(y,x,'b')

axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Graph 1')

axes2.set_xlabel('Y')
axes2.set_ylabel('X')
axes2.set_title('Graph 2')
```
