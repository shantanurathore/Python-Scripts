```python
%matplotlib inline
from pylab import *

import matplotlib.pyplot as plt
import numpy as np

#np.linspace probably creates a list
x=np.linspace(0,500,10) #the 10 here is probably log to the base 10
y=x**2

figure()
plot(x,y,'r') #r stands for color red g is green b is blue
xlabel('x')
ylabel('y')
title('square function')
show()

```

```python
#subplots -- what are subplots? I think they are layout managers in matplotlib
subplot(1,2,1)
plot(x, y, 'r--') # r-- means red line of type ----
subplot(1,2,2)
plot(y, x, 'g*-'); # g*- means green line with * and - 
```

```python
#defining axes - location of the axes for a plot
#fig here is a global variable which can support multiple plots and plots within plots
#its the OOP feature of matplotlib package
fig = plt.figure()
axes = fig.add_axes([0.1,0.1,0.8,0.8]) #left, bottom, width, height - values are between 0 and 1
#0.1 is basically start of the plot
axes.plot(x,y,'r')

axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('TITLE')
```

```python
#example of adding multiple axes to the plot - two graphs in a single space
fig = plt.figure()
#axes1 is the first graph
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])#left, bottom, width, height - values are between 0 and 1
#0.1 is basically start of the plot
#axes2 is the second graph within the first graph
axes2 = fig.add_axes([0.15,0.39,0.4,0.4])

axes1.plot(x,y,'r')
axes2.plot(y,x,'b')

axes1.set_xlabel('x')
axes1.set_ylabel('y')
axes1.set_title('Graph 1')

axes2.set_xlabel('Y')
axes2.set_ylabel('X')
axes2.set_title('Graph 2')
```

```python
#SUBPLOTS - managing axes using matplotlib without the need to define the plot position (0.1 to 0.8 etc)
#the simple use of subplots
fig, axes = plt.subplots()

axes.plot(x,y,'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('Graph')
```

```python
#the more complex use of subplots - using a tabular structure for our graphs
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x,y,'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('simple graph')
```

```python
#as we can see in previous figure, labels and markings overlap. We can fix it as follows
fig, axes = plt.subplots(nrows=1, ncols=2)

for ax in axes:
    ax.plot(x,y,'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('simple graph')
    
fig.tight_layout()
```

```python
#we can cahnge DPI and width of graphs as well as follows
fig, axes = plt.subplots(figsize=(10,10), dpi=100)

axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('large graph');
```

```python
#saving the file in the root directory
fig.savefig("graphs.png", dpi=200)
```
